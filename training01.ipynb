```python
import os
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data=pd.read_csv('/content/sample_data/black_friday_data_extra_full.csv')
```

```python
data.head(5)
```

```python
data.describe()
```

```python
commentaires=data["Comment"]
data.drop(columns=["Comment"],inplace=True)
```

```python
commentaires
```

```python
# Variables pour compter les sentiments
positifs = 0
negatifs = 0
neutres = 0
```

```python
# Analyse de sentiment pour chaque avis
for commentaire in commentaires:
    blob = TextBlob(commentaire)
    polarite = blob.sentiment.polarity

     # Classification des avis en fonction de la polarité
    if polarite > 0:
        positifs += 1
    elif polarite < 0:
        negatifs += 1
    else:
        neutres += 1
```

```python
# Données pour le graphique
categories = ['Positifs', 'Négatifs', 'Neutres']
nombre_avis = [positifs, negatifs, neutres]
```

```python
# Création du graphique à barres
plt.figure(figsize=(8, 6))
plt.bar(categories, nombre_avis, color=['green', 'red', 'gray'])

# Titre et labels
plt.title('Analyse des sentiments des avis de consommateurs', fontsize=16)
plt.xlabel('Type de sentiment', fontsize=14)
plt.ylabel('Nombre d\'avis', fontsize=14)

# Sauvegarde du graphique dans le même dossier que le script
# plt.savefig('graphique_analyse_sentiment.png')

# Afficher le graphique
plt.show()
```

```python
# Initialiser le style de Seaborn pour des graphiques esthétiques
sns.set(style="whitegrid")

# Taille de la figure
plt.figure(figsize=(10, 6))

# Créer un nuage de points (scatter plot) pour comparer Price Before et Price After
sns.scatterplot(x='Price Before', y='Price After', data=data, alpha=0.5)

# Ajouter une ligne diagonale (où le prix avant et après sont identiques, pour visualiser les faux rabais)
max_price = max(data['Price Before'].min(), data['Price After'].min())  # Trouver le maximum des prix
plt.plot([0, max_price], [0, max_price], color='red', linestyle='--', label="No Discount (Fraudulent)")

# Ajouter des titres et labels aux axes
plt.title('Comparison of Price Before and After Black Friday', fontsize=16)
plt.xlabel('Price Before Black Friday', fontsize=12)
plt.ylabel('Price After Black Friday', fontsize=12)

# Ajouter une légende pour la ligne diagonale
plt.legend()

# Afficher le graphique
plt.show()
```
